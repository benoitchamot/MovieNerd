# Data Exploration
## Questions
1. What is the average rating for each actor?
2. Who is the best director (based on ratings) per genre?


```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Import datasets
omdb_csv = Path('MONU_project3/Datasets/omdb.csv')
omdb_df = pd.read_csv(omdb_csv)
omdb_df.head()
```

```python
omdb_df.columns
```

```python
omdb_df = omdb_df[['Title','Genre','Director','Actors', 'imdbRating', 'imdbVotes' ]]
omdb_df
```

```python
# Splitting the Actors column into separate actors
omdb_df[["Actor_1","Actor_2", "Actor_3"]] = omdb_df["Actors"].str.split(', ', n=2, expand=True)
omdb_df
```

```python
# Stacking columns 'Actor_1', 'Actor_2', 'Actor_3' into one column
actors_df = omdb_df.set_index(['Title', 'Genre', 'Director', 'Actors', 'imdbRating', 'imdbVotes'])[['Actor_1', 'Actor_2', 'Actor_3']].stack().reset_index()

# Renaming the column as "Actor"
actors_df = actors_df.rename(columns={0: 'Actor'})

# Dropping column 'level_6'
actors_df = actors_df.drop('level_6', axis=1)

actors_df
```

```python
# actors_df.to_csv("actors.csv", index = False, header = True)
```

```python
# Calculating the average rating per actor (using groupby method)
actor_avg_rating = actors_df[['Actor','imdbRating']]
actor_avg_rating = actor_avg_rating.groupby('Actor')
actor_avg_rating = actor_avg_rating.mean().reset_index()
actor_avg_rating 
```

```python
# actor_avg_rating.to_csv("rating.csv", index = False, header = True)
```

```python
# Top 10 actors by rating
top_10_actors_rating = actor_avg_rating.sort_values("imdbRating", ascending = False)
top_10_actors_rating = top_10_actors_rating.head(10)
top_10_actors_rating
```

```python
# actor_avg_rating.to_csv("file4.csv", index = False, header = True)
```

```python
# Calculating the number of movies an actor played in and selecting top 10 actors who played in the biggest amount of movies
actor_count_movies = actors_df['Actor'].value_counts()

actor_count_movies
```

```python
top_10_actors_movies = actor_count_movies.head(10)
top_10_actors_movies = pd.DataFrame(top_10_actors_movies)
top_10_actors_movies.reset_index()

top_10_actors_movies = top_10_actors_movies.rename(columns={"Actor": "Count of movies"})

top_10_actors_movies = top_10_actors_movies.rename_axis('Actor')
top_10_actors_movies
```

```python

actor_count_movies = pd.DataFrame(actor_count_movies)
actor_count_movies  = actor_count_movies.rename(columns={"Actor": "Count of movies"})
actor_count_movies  = actor_count_movies.rename_axis('Actor')
actor_count_movies

```

```python

top_10_actors_movies_ratings = pd.merge(top_10_actors_rating, actor_count_movies , on="Actor")
top_10_actors_movies_ratings
```

```python
# Comparison between most prolific actors and their ratings
# Of the top 10 actors who have acted in the most movies, how do their average ratings compare? 

top_10_actors_ratings_movies = pd.merge(top_10_actors_movies, actor_avg_rating , on="Actor")
top_10_actors_ratings_movies
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a histogram
plt.hist(actor_avg_rating ['imdbRating'], bins=20, color='#1FCBB8')

# Add a title and labels
plt.title('Distribution of Average Ratings for Actors')
plt.xlabel('IMDB Rating')
plt.ylabel('Number of Actors')
plt.grid()

# Show the plot
plt.show()
```

```python


bar_chart = top_10_actors_rating.set_index('Actor')
bar_chart 

bars = plt.bar(bar_chart.index, bar_chart ['imdbRating'])
plt.title('Average IMDB Ratings for Top Actors')
plt.xlabel('Actors')
plt.ylabel('IMDB Rating')
plt.xticks(rotation=45, ha='right')
# Annotate each bar with its value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval,2), ha='center', va='bottom')

    
plt.grid()
plt.show()
```

```python
# To visualize if there's a correlation between the number of movies an actor has done and their average rating
actor_avg_rating

actor_count_movies = actor_count_movies.reset_index()
# actor_count_movies

# Merging the two DataFrames on the 'Actor' column
merged_df = pd.merge(actor_count_movies, actor_avg_rating, on='Actor')
merged_df
```
